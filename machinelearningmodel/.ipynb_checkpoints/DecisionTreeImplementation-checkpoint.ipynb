```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
```

```python
df = pd.read_csv("time_series_dump_v1.csv")
df.head(5)
```

```python
#Checking the column list
print(list(df))
```

```python
#Checking the datatype of columns
df['sleepHours'].dtypes
```

```python
#Eliminating the unnecessary columns
df = df[['stepCount(cumulativeforday)', 'bloodOxygenAvg(mmHg)', 
         'stress(0to5)', 'sleepHours', 'REMsleepHours', 
         'caloriesBurnt(cumulativeforday)', 'heartRate(bpm)', 
         'systolicBloodPressure(mmHg)', 'diastolicBloodPressure(mmHg)', 
         'bodyTemperature(F)', 'breathingFrequency(perminute)', 'healthState']]
```

```python
from sklearn.tree import DecisionTreeClassifier 
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
#Splitting in features and target variable
feature_cols = ['stepCount(cumulativeforday)', 'bloodOxygenAvg(mmHg)', 
         'stress(0to5)', 'sleepHours', 'REMsleepHours', 
         'caloriesBurnt(cumulativeforday)', 'heartRate(bpm)', 
         'systolicBloodPressure(mmHg)', 'diastolicBloodPressure(mmHg)', 
         'bodyTemperature(F)', 'breathingFrequency(perminute)']
x = df[feature_cols] # Features
y = df.healthState # Target variable
```

```python
#Train-test splitt

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1) # 70% training and 30% test
```

```python
# Create Decision Tree classifier object
clf = DecisionTreeClassifier(criterion = "entropy", max_depth = 8, random_state = 15, min_samples_split=2)
```

```python
# Train Decision Tree Classifier
clf = clf.fit(x_train,y_train)
```

```python
#Predict the response for test dataset
y_pred = clf.predict(x_test)
y_pred
```

```python
clf.predict([[1375,95,4,8,3,1221,148,70,120,97,17]])
```

```python
clf.predict([[987,95,4,8,2,1158,130,80,120,97,15]])
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score

# calculate the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print('Confusion Matrix:', conf_matrix)

# calculate the classification report
class_report = classification_report(y_test, y_pred)
print('Classification Report:', class_report)

# calculate the AUC-ROC score
#roc_auc = roc_auc_score(y_test, y_pred)
#print('AUC-ROC:', roc_auc)
```

```python
from sklearn.tree import export_graphviz
from six import StringIO  
from IPython.display import Image  
import graphviz

import pydotplus
dot_data = StringIO()
export_graphviz(clf, out_file = dot_data,  
                filled = True, rounded = True,
                special_characters = True, feature_names = feature_cols, class_names = ['0','1'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png('Healthstate.png')
Image(graph.create_png())
```

```python
import pickle

# Train your ML model
model = clf # Trained ML model

# Save the model to disk using pickle
with open("model.pkl_0.0", "wb") as file:
    pickle.dump(model, file)
```
